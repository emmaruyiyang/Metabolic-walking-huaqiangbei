### Human Shade Detection

This code can:
- Person Tracking: Capture coordinate information for each person in every frame
- Shade Detection: Identify shadow regions within the image
- Person-Shade Relationship Analysis: Determine whether individuals are positioned within shadowed areas

Input: preprocessed video

Output: csv and annotated video

```python
from collections import defaultdict

import cv2
import numpy as np

from ultralytics import YOLO

# Load the YOLO26 model
model = YOLO("yolo26n.pt")

# Open the video file
cap = cv2.VideoCapture(video_path)

# Store the track history
track_history = defaultdict(lambda: [])

box_color = (255, 255, 255)  # White color for the box
line_color = (255, 255, 255)  # White color for the line
line_thickness = 2

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # 先添加灰色半透明蒙版到原始frame
        gray_mask = np.zeros_like(frame) * 100  # 灰色值，zeros_like或者ones_like
        frame = cv2.addWeighted(frame, 0.5, gray_mask, 0.5, 0)
        # frame - 原始画面
        # 0.7 - 原始画面的权重（70%）
        # gray_mask - 灰色蒙版
        # 0.3 - 蒙版的权重（30%）
        # 0 - 亮度偏移
        
        # Run YOLO26 tracking on the frame, persisting tracks between frames
        result = model.track(frame, 
                             conf=0.1,  # 置信度阈值
                             classes=[0],  # 只track人类（class 0）
                             tracker=tracker,
                             persist=True)[0]

        # Get the boxes and track IDs
        if result.boxes and result.boxes.is_track:
            boxes = result.boxes.xyxy.cpu()  # 获取xyxy格式的边界框
            track_ids = result.boxes.id.int().cpu().tolist()

            # Plot the tracks with custom styling
            for box, track_id in zip(boxes, track_ids):
                x1, y1, x2, y2 = box
                
                # Draw bounding box with white color and thin line
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), box_color, line_thickness)
                
                # Draw track ID label
                cv2.putText(frame, f"ID {track_id}", (int(x1), int(y1) - 5), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, box_color, line_thickness)
                
                # Store center point for tracking
                center_x = (x1 + x2) / 2
                center_y = (y1 + y2) / 2
                track = track_history[track_id]
                track.append((float(center_x), float(center_y)))
                
                if len(track) > 30:  # retain 30 tracks for 30 frames
                    track.pop(0)

                # Draw the tracking lines with white color (无论track长度如何都绘制)
                if len(track) >= 1:
                    points = np.hstack(track).astype(np.int32).reshape((-1, 1, 2))
                    cv2.polylines(frame, [points], isClosed=False, color=line_color, thickness=2)

        # Display the annotated frame
        cv2.imshow("YOLO26 Tracking", frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

```python
from collections import defaultdict

import cv2
import numpy as np

from ultralytics import YOLO

# 定义函数：检测是否为外卖小哥
def is_delivery_person(box, frame, orange_threshold=0.05, backpack_threshold=0.3):
    """
    检测一个人是否为外卖小哥
    
    参数:
    - box: [x1, y1, x2, y2] 边界框坐标
    - frame: 原始视频帧
    - orange_threshold: 橙色区域占比阈值（0-1）
    - backpack_threshold: 背包特征阈值
    
    返回:
    - is_delivery: Boolean
    - confidence: 0-1 置信度
    """
    x1, y1, x2, y2 = int(box[0]), int(box[1]), int(box[2]), int(box[3])
    
    # 提取人物ROI区域
    roi = frame[y1:y2, x1:x2]
    
    if roi.size == 0:
        return False, 0
    
    h, w = roi.shape[:2]
    
    # 特征1: 检测橙色/黄色衣物（骑手服装特征）
    hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
    
    # 橙色范围 (HSV)
    orange_lower = np.array([5, 100, 100])
    orange_upper = np.array([25, 255, 255])
    orange_mask = cv2.inRange(hsv, orange_lower, orange_upper)
    
    # 黄色范围 (HSV)
    yellow_lower = np.array([20, 100, 100])
    yellow_upper = np.array([35, 255, 255])
    yellow_mask = cv2.inRange(hsv, yellow_lower, yellow_upper)
    
    # 合并橙黄色掩码
    color_mask = cv2.bitwise_or(orange_mask, yellow_mask)
    orange_ratio = color_mask.sum() / (h * w * 255)
    
    # 特征2: 检测背包（上半身有大面积较暗区域）
    upper_half = roi[:h//2, :]
    lower_half = roi[h//2:, :]
    
    # 计算亮度差异
    upper_brightness = cv2.cvtColor(upper_half, cv2.COLOR_BGR2GRAY).mean()
    lower_brightness = cv2.cvtColor(lower_half, cv2.COLOR_BGR2GRAY).mean()
    
    # 背包通常上半身较暗
    backpack_score = max(0, (lower_brightness - upper_brightness) / 255.0)
    
    # 综合判断
    delivery_score = orange_ratio * 0.6 + backpack_score * 0.4
    
    is_delivery = delivery_score > 0.15  # 阈值
    
    return is_delivery, delivery_score

# Load the YOLO model
model = YOLO("yolo26n.pt")

# Open the video file
cap = cv2.VideoCapture(video_path)

# Store the track history
track_history = defaultdict(lambda: [])
person_type = defaultdict(lambda: None)  # 存储每个track_id的人员类型

line_thickness = 1

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # 先添加灰色半透明蒙版到原始frame
        gray_mask = np.ones_like(frame) * 100  # 灰色值
        frame = cv2.addWeighted(frame, 0.7, gray_mask, 0.3, 0)
        
        # Run YOLO tracking on the frame
        result = model.track(frame, 
                             conf=0.1,
                             classes=[0],
                             tracker=tracker,
                             persist=True)[0]

        # Get the boxes and track IDs
        if result.boxes and result.boxes.is_track:
            boxes = result.boxes.xyxy.cpu()
            track_ids = result.boxes.id.int().cpu().tolist()

            # Plot the tracks with custom styling
            for box, track_id in zip(boxes, track_ids):
                x1, y1, x2, y2 = box
                
                # 检测是否为外卖小哥
                is_delivery, confidence = is_delivery_person(box, frame)
                person_type[track_id] = (is_delivery, confidence)
                
                # 根据类型选择颜色
                if is_delivery:
                    box_color = (0, 165, 255)  # 橙色 (BGR)
                    label_text = f"ID {track_id} [Delivery] {confidence:.2f}"
                else:
                    box_color = (255, 255, 255)  # 白色
                    label_text = f"ID {track_id} [Person]"
                
                # Draw bounding box
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), box_color, line_thickness)
                
                # Draw track ID label
                cv2.putText(frame, label_text, (int(x1), int(y1) - 5), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, box_color, line_thickness)
                
                # Store center point for tracking
                center_x = (x1 + x2) / 2
                center_y = (y1 + y2) / 2
                track = track_history[track_id]
                track.append((float(center_x), float(center_y)))
                
                if len(track) > 30:
                    track.pop(0)

                # Draw the tracking lines
                if len(track) >= 1:
                    points = np.hstack(track).astype(np.int32).reshape((-1, 1, 2))
                    cv2.polylines(frame, [points], isClosed=False, color=box_color, thickness=2)

        # Display the annotated frame
        cv2.imshow("YOLO Tracking - Delivery Person Detection", frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()

print("Tracking complete!")
```

```python
'''
function: detect people whether in the shadow from the video
change the parameters: skip
'''
skip = 2  # Set the number of frame skips

output_dir = "output"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    
output_video_path = "output/human-and-shade-tracking_{}.avi".format(file_name)
csv_file = 'output/detections_{}.csv'.format(file_name)

# Initialize Roboflow client for shadow detection
CLIENT = InferenceHTTPClient(api_url="https://detect.roboflow.com", api_key="xxx")

# Load the YOLO model for person detection
model_person = YOLO('yolo11l.pt').to(device)

# Open the video file
cap = cv2.VideoCapture(video_path)
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*"MJPG"), fps, (w, h))

# Initialize CSV file
csv_headers = ['frame', 'person_id', 'bottom_mid_x', 'bottom_mid_y', 'in_shadow']
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(csv_headers)

frame_count = 0

# Statistics variables
total_people = 0
people_in_shadow = 0
people_outside_shadow = 0

# Create a window and set its size
# cv2.namedWindow("instance-segmentation-object-tracking", cv2.WINDOW_NORMAL)
# cv2.resizeWindow("instance-segmentation-object-tracking", 800, 600)

# Start processing each frame
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break

    frame_count += 1

    # Perform people count detection every frame, change here
    if frame_count % skip == 0:
        # Get the shadow boundary of the current frame
        shadow_result = CLIENT.infer(frame, model_id="shade-detection/1")
        boundaries = []

        # Data structure for single image inference
        if 'predictions' in shadow_result: 
            predictions = shadow_result['predictions']
            for prediction in predictions:
                points = prediction.get("points", [])
                if points:
                    boundary = [(int(point['x']), int(point['y'])) for point in points]
                    boundaries.append(boundary)  # Add the point coordinates of the shadow area to the boundaries list

        annotator = Annotator(frame, line_width=2)
        
        # Reset statistics variables
        people_in_shadow = 0
        people_outside_shadow = 0
        total_people = 0
        detections = [] 

        # Draw the shadow polygon
        for boundary in boundaries:
            boundary_points = np.array(boundary, np.int32).reshape((-1, 1, 2))
            cv2.polylines(frame, [boundary_points], isClosed=True, color=(255, 0, 0), thickness=2)

        # people detection
        # person_results = model_person(frame, classes=[0], conf=0.05, show=False)  
        person_results = model_person.track(frame, persist=True, classes=[0], conf=0.05, show=False) # 0 is the category ID of the person, close the automatic display window
        
        if person_results and len(person_results) > 0:
            boxes = person_results[0].boxes

            if boxes and len(boxes) > 0:
                # Extract bounding boxes
                xyxy = boxes.xyxy.cpu().numpy()  # Get the box in xyxy format
                track_ids = np.arange(len(xyxy))  # Generate a unique ID for each box

                for i, (x1, y1, x2, y2) in enumerate(xyxy):
                    bottom_mid_x = (x1 + x2) / 2  # X coordinate of the midpoint of the bottom edge
                    bottom_mid_y = y2  # Y coordinate of the bottom edge
                    in_shadow = False

                    # Check if in shadow
                    for boundary in boundaries:
                        if cv2.pointPolygonTest(np.array(boundary, np.int32), (int(bottom_mid_x), int(bottom_mid_y)), False) >= 0:
                            in_shadow = True
                            people_in_shadow += 1
                            break
                    else:
                        people_outside_shadow += 1

                    # Record test results
                    detections.append([frame_count, track_ids[i], bottom_mid_x, bottom_mid_y, in_shadow])
                    total_people += 1

                    # Display the detection box, text and the dot at the bottom midpoint
                    color = colors(int(track_ids[i]), True)
                    txt_color = annotator.get_txt_color(color)
                    label = f"ID {track_ids[i]} {'in shadow' if in_shadow else 'out shadow'}"
                    annotator.box_label([x1, y1, x2, y2], label=label, color=color, txt_color=txt_color)
                    cv2.circle(frame, (int(bottom_mid_x), int(bottom_mid_y)), 5, (0, 0, 255), -1)  

        # Write the results to CSV
        with open(csv_file, 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(detections)

        # Display the statistics results on the frame
        text = f"Frame {frame_count}: Total: {total_people}, In Shadow: {people_in_shadow}, Outside Shadow: {people_outside_shadow}"
        cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

        out.write(frame)
        cv2.imshow("instance-segmentation-object-tracking", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Processing complete. Results saved in {output_video_path} and {csv_file}.")
```
